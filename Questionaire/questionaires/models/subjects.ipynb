```python
import pandas as pd
```

```python
df = pd.read_csv("D:/Datasets/subjects/train.csv")
df
```

```python
df.drop("Id", axis="columns", inplace=True)
```

```python
df
```

```python
df.Topic.unique()
```

```python
df.Topic = df.Topic.replace(["Biology", "Physics", "Chemistry"], [0, 1, 2])
```

```python
df
```

```python
df.Comment[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = df.Comment
labels = df.Topic
```

```python
labels.dtype
```

```python
vect = TfidfVectorizer()
```

```python
vect_data = vect.fit_transform(data)
```

```python
vect.get_feature_names_out()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.pipeline import Pipeline
```

```python
model = Pipeline([
    ("vectorizer", TfidfVectorizer()),
    ("classifier", RandomForestClassifier())
])
```

```python
model.fit(data, labels)
```

```python
test_data = df = pd.read_csv("D:/Datasets/subjects/test.csv")
```

```python
test_data
```

```python
test_data.Topic = test_data.Topic.replace(["Biology", "Physics", "Chemistry"], [0, 1, 2])
```

```python
test_data
```

```python
test_data_classification = test_data.Comment
test_labels = test_data.Topic
```

```python
check_data = model.predict(test_data_classification)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(check_data, test_labels)
```

```python
score
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mb_model = Pipeline([
    ("vectorizer", TfidfVectorizer()),
    ("model", MultinomialNB(alpha=0.8, fit_prior=True, force_alpha=True))
])
```

```python
mb_model.fit(data, labels)
```

```python
score_2 = mb_model.predict(test_data_classification)
```

```python
accuracy_score(score_2, test_labels)
```

```python
import joblib
```

```python
joblib.dump(mb_model, "D:/Datasets/models.joblib")
```

```python
model_2 = joblib.load("D:/Datasets/models.joblib")
```

```python
score_3 = model_2.predict(test_data_classification)
```
